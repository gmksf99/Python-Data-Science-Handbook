# Matplotlib
- 매직 명령어
  - %matplotlib notebook : 대화형 플롯을 삽입
  - %matplotlib inline : 플롯의 정적이미지를 삽입

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#linspace(시작,끝,개수) : 일정 간격으로 1차원 배열 만들기
x = np.linspace(0, 10, 100)

# plot(x축, y축)
fig = plt.figure()
plt.plot(x, np.sin(x), '-') # 실선
plt.plot(x, np.cos(x), '--') # 점선
```

```
# 그림으로 파일 저장
fig.savefig('my_figure.png')

# 디렉토리에 있는 png파일을 출력_ 기본 플롯의 PNG 버전
from IPython.display import Image
Image('my_figure.png')
```

# Matplot이 제공하는 인터페이스
- 매트랩 스타일으 상태기반 인터페이스
- 더 강력한 객체지향 인터페이스

```
# 매트랩 스타일의 인터페이스

plt.figure() # 플롯 그림 생성

# 두개의 패널 중 첫 번째 패널을 생성하고 현재 축(axis)을 설정
plt.subplot(2,1,1) #(rows, columns, panel number)
plt.plot(x,np.sin(x))

# 두개의 패널 중 두 번째 패널을 생성하고 현재 축(axis)을 설정
plt.subplot(2,1,2) 
plt.plot(x,np.cos(x))
```

```
# 객체지향 인터페이스_ 그림을 제어하고 싶을 때

# 먼저 플롯 그리드를 생성
# ax(축)는 그래프를 생성할 틀의 개수라 생각
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

# 산점도

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 스타일 설정_ 그리드

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```
# 선스타일처럼 표시자 스타일도 있다
# RandomState : 선언시에 각 변수 마다 다른 값을 넣음으로서 하나의 코드상에서 다른 랜덤한 값을 출력할 수 있을
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
  # format : 문자열 중간 중간에 특정 변수의 값을 넣어주기 위해서 사용
  # '{인덱스0}, {인덱스1}'.format(값0, 값1)
  plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
  plt.legend(numpoints=1)
  plt.xlim(0, 1.8)
```

```
plt.plot(x,y,'-ok') # 선(-), 원 기호(o), 블랙(k)
```

```
rng = np.random.RandomState(123)

#randn() : 평균0, 표준편차1의 가우시안 표준정규분포 난수 생성
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

# alpth : 투명도
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 색상 척도 표시
```

# 오차 시각화

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 스타일 설정_ 그리드

x = np.linspace(0,10,30)
dy = 0.5 # 데이터의 편차
y = np.sin(x) + dy * np.random.randn(30)

plt.errorbar(x,y,yerr=dy, fmt='.')
```

```
plt.errorbar(x,y,yerr=dy, fmt='o', color='k', ecolor='lightgray', elinewidth=3, capsize=0)
```

# 밀도 플롯과 등고선 플롯
- 3차원 데이터를 2차원으로 표시하는데 유용

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') 

# 3차원 함수 시각화
def f(x,y) :
  return np.sin(x) ** 10 + np.cos(10+y*x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

# emshgrid() : 2차원 그리드 좌표를 반환
X, Y = np.meshgrid(x,y)
Z = f(X,Y)

# 단색을 사용하면 음수 값은 점선 / 양수 값은 실선으로 표시된다
plt.contour(X, Y, Z, colors='k')
```

```
# cmap : 색상표를 지정해 선에 색을 입힐 수 있음
# 20 : 데이터 범위내에서 20개의 동일한 간격으로 더 많은 선을 그림
plt.contour(X, Y, Z, 20, cmap='RdGy') #(Red-Gray 약어)
```

```
# 선이 아닌 색으로 채워진 등고선 플롯
plt.contourf(X, Y, Z, 20, cmap='RdGy')
# 검은색은 높은 지정, 빨간색은 낮은 지점을 의미
plt.colorbar()
```

# 히스토그램, 구간화, 밀도

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 스타일 설정_ 그리드

data = np.random.randn(1000)

plt.hist(data)
```

```
# bins : 막대영역을 얼마나 쪼갤지
plt.hist(data, bins=30, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='green')
```

```
# alpha를 주어 그래프들을 겹치게 만들 수 있음
# random.normal() : 정규 분포 (Normal distribution)로부터 샘플링된 난수를 반환
x1 = np.random.normal(0, 0.8, 1000) # N(0, 0.8^2)로 부터 얻은 1000개의 임의 숫자
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

# 그래프 속성 설정
kwargs = dict(bins=40, alpha=0.3, histtype='stepfilled')

plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
```

# 플롯 범례 맞춤 변경
- 범례의 모샹, 위치를 맞춤 변경

```
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np

x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), '--', label='Cosine')
ax.axis('equal')
leg = ax.legend()
```

```
# 범례 위치 및 테두리 없애기
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), '--', label='Cosine')
ax.axis('equal')
ax.legend(loc='upper left', frameon=False)
```

```
# 범례에 사용되는 열의 개수 지정 : ncol 명령어
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), '--', label='Cosine')
ax.axis('equal')
ax.legend(loc='lower center', frameon=False, ncol=2)
```

```
# 다중 범례
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0,10,1000)

for i in range(4) :
  lines += ax.plot(x, np.sin(x-i * np.pi/2), styles[i], color='k')
ax.axis('equal')

# 첫 번째 범례의 라인과 레이블 지정
ax.legend(lines[:2], ['line A', 'line B'], loc='upper right', frameon=False)

# 두 번째 범례의 라인과 레이블 지정
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'], loc='lower right', frameon=False)
ax.add_artist(leg)
```
